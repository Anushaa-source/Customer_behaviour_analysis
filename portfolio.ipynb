```python
#to replace the missing values iwth median

#df=df.fillna (df.median(numeric_only=True))
#for a single  column- df['purchase'] = df['purchase'].fillna(df['purchase'].median())
# the best methos is 
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))


```

```python
df.isnull().sum()
```

```python
#formatting its beteter practise to convert column name to snake casiuing for esay to read, consistency,reference

df.columns = df.columns.str.lower()
df.columns=df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a new column age group based on a condition
labels=['young','middle-age','young adult','senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(5)
```

```python
#craeate anotehr frequency purchase_frequency_days corersponding “The purchase frequency column was in string format (like Weekly, Fortnightly).
#To use it for analysis, I created a mapping dictionary that assigns a numeric day equivalent to each frequency.
#Then I used .map() to convert the entire categorical column into a numeric column called purchase_frequency_days.
#This made the data consistent and ready for analysis.”

freq_map = {
    'Weekly': 7,
    'Bi-Weekly': 14,
    'Fortnightly': 14,      
    'Monthly': 30,
    'Quarterly': 90,
    'Annually': 365,
    'Every 3 months':90
}

df['purchase_frequency_days']=df['frequency_of_purchases'].map(freq_map)
```
